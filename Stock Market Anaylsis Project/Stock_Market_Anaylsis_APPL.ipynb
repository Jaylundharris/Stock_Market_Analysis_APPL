```python
import pandas as pd
print(pd.__version__)

import numpy as np
print(np.__version__)


import seaborn as sns
print(sns.__version__)

import matplotlib
import yfinance
print("matplotlib:", matplotlib.__version__)
print("yfinance:", yfinance.__version__)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import seaborn as sns

plt.rcParams['figure.figsize'] = (10, 5)
print("âœ… Libraries ready:")
print("pandas:", pd.__version__)
print("matplotlib:", plt.matplotlib.__version__)
print("yfinance:", yf.__version__)

data = yf.download("AAPL", start="2022-01-01", end="2024-12-31")
data.head()

plt.plot(data.index, data["Close"], label="Close Price")
plt.title("AAPL Closing Prices")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.legend()
plt.show()
```

### ðŸ“Š Quick Insight
- Appleâ€™s stock has trended upward since early 2023 with noticeable dips around major earnings.
- Next step: compare moving averages (20 vs 50 day) to see how trend strength changes.

```python
data["SMA20"] = data["Close"].rolling(20).mean()
data["SMA50"] = data["Close"].rolling(50).mean()

data[["Close", "SMA20", "SMA50"]].plot(title="AAPL Close vs SMA20 & SMA50")
plt.show()
```
